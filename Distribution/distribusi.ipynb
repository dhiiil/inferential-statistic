```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Definisikan jumlah sampel
samples = 500
```

```python
def create_frequency_table(data):
    df = pd.DataFrame(data, columns=['values'])
    df['Binned'] = pd.cut(df['values'], bins=10)
    frequency_table = df['Binned'].value_counts().sort_index()
    frequency_df = frequency_table.reset_index()
    frequency_df.columns = ['Interval', 'Frequency']

    return frequency_df

def hist_plot_visualize(data):
    plt.figure(figsize=(12,4))
    sns.histplot(data, kde=True)
    plt.grid(True)
    plt.title("Visualisasi Distribusi dari 500 Sampel Acak")
```

# 1. Uniform Distribution

```python
uniform_distributed = np.random.rand(samples)
```

```python
uniform_distributed
```

```python
frequency_table = create_frequency_table(uniform_distributed)
frequency_table
```

```python
hist_plot_visualize(uniform_distributed)
```

# 2. Normal Distribution

```python
standard_normal_distributed = np.random.randn(samples)
```

```python
frequency_table = create_frequency_table(standard_normal_distributed)
frequency_table
```

```python
hist_plot_visualize(standard_normal_distributed)
```

# 3. Poisson Distribution

```python
poisson_distributed = np.random.poisson(2, samples)
```

```python
frequency_table = create_frequency_table(poisson_distributed)
frequency_table
```

```python
hist_plot_visualize(poisson_distributed)
```

# 4. Exponential Distribution

```python
exponential_distributed = np.random.power(5, samples)
```

```python
frequency_table = create_frequency_table(exponential_distributed)
frequency_table
```

```python
hist_plot_visualize(exponential_distributed)
```

# 5. Distribusi Binomial

```python
n, p = 10, .5
binomial_distributed = np.random.binomial(n, p, 1000)
```

```python
frequency_table = create_frequency_table(binomial_distributed)
frequency_table
```

```python
hist_plot_visualize(exponential_distributed)
```
